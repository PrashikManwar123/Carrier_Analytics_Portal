```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('HR_Analytics.csv')
df
```

```python
df2 = df[df['JobLevel']==5]
print(df2["YearsWithCurrManager"].value_counts())
```

```python
df.isnull().sum()
#checking for null values
```

```python
#dropping rows with null values
df = df.dropna()
df #this is the cleaned dataset
```

```python
# Define key features and target
df = df[['JobLevel','JobRole','TotalWorkingYears','Department','MonthlyIncome']]

X = df.drop(columns=['MonthlyIncome']) #Key features 
y = df[['MonthlyIncome']] #target
```

```python
df['JobRole'].value_counts()
```

```python
df['JobLevel'].value_counts()
```

```python
df.tail()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['JobRole'] = le.fit_transform(df['JobRole'])
df['Department'] = le.fit_transform(df['Department'])
df['JobRole'].value_counts()
```

```python
df['Department'].value_counts()
```

```python
df.tail()
#this is the final dataset
```

```python
#Predicting using CatBoost
from catboost import CatBoostRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Identify categorical features (CatBoost will handle them automatically)
categorical_features = ['JobRole', 'Department']

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Initialize CatBoostRegressor
model = CatBoostRegressor(
    cat_features=categorical_features,  # Specify categorical columns
    iterations=1000,                   # Number of boosting iterations
    learning_rate=0.01,                 # Learning rate
    depth=6,                           # Tree depth
    loss_function='RMSE',              # Regression loss (Root Mean Squared Error)
    verbose=100                        # Print training progress every 100 iterations
)
```

```python
# Train the model
model.fit(X_train, y_train, eval_set=(X_test, y_test), plot=True)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
# Evaluate performance
print(f"RMSE: {mean_squared_error(y_test, y_pred, squared=False)}")
print(f"R² Score: {r2_score(y_test, y_pred)}")
```

```python
# Feature importance
feature_importance = model.get_feature_importance()
feature_names = X_train.columns
for score, name in sorted(zip(feature_importance, feature_names), reverse=True):
    print(f"{name}: {score}")
```
